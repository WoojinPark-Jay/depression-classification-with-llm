```python
import pandas as pd
```

```python
simple_depression_data = pd.read_csv("/Users/hany/Documents/KU-NLP/new_simple_depression_data.csv")
simple_neutral_data = pd.read_csv("/Users/hany/Documents/KU-NLP/new_simple_neutral_data.csv")
simple_happy_data = pd.read_csv("/Users/hany/Documents/KU-NLP/new_simple_happy_data.csv")
```

```python
Depression_Group_df = simple_depression_data[simple_depression_data['polarity']< 0.0]
Neutral_Group_df =simple_neutral_data.loc[(simple_neutral_data['polarity'] > -0.2) & (simple_neutral_data['polarity'] < 0.2)]
Happy_Group_df = simple_happy_data[simple_happy_data['polarity'] > 0.0]
```

```python
print(len(Depression_Group_df))
print(len(Neutral_Group_df))
print(len(Happy_Group_df))
```

```python
neutral_short_df = Neutral_Group_df[Neutral_Group_df['cleaned_length']<4]
len(neutral_short_df)
```

```python
happy_short_df = Happy_Group_df[Happy_Group_df['cleaned_length']<3]
len(happy_short_df)
```

```python
depression_short_df = Depression_Group_df[Depression_Group_df['cleaned_length']<4]
len(depression_short_df)
```

```python
depression_mid_df = Depression_Group_df[(Depression_Group_df['cleaned_length']>4) & (Depression_Group_df['cleaned_length']<6)]
len(depression_mid_df)
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import pyplot

import numpy as np
```

```python
plt.figure(figsize=(20,10))
bins = np.linspace(-1, 1, 300)

pyplot.hist(happy_short_df['new_polarity'], bins, alpha=0.5, label='short_depression',color='red')

plt.xlabel('new_polarity',fontsize =20)
plt.ylabel('Count',fontsize =20)
pyplot.legend(loc='upper right',fontsize=20)
plt.title('Happy',fontsize =20)
```

```python
plt.figure(figsize=(20,10))
bins = np.linspace(-1, 1, 300)

pyplot.hist(depression_short_df['new_polarity'], bins, alpha=0.5, label='short_depression',color='red')
pyplot.hist(depression_mid_df['new_polarity'], bins, alpha=0.5, label='mid_depression',color='green')


plt.xlabel('new_polarity',fontsize =20)
plt.ylabel('Count',fontsize =20)
pyplot.legend(loc='upper right',fontsize=20)
plt.title('depression',fontsize =20)
```

```python
plt.figure(figsize=(20,10))
bins = np.linspace(-1, 1, 300)

pyplot.hist(neutral_short_df['new_polarity'], bins, alpha=0.5, label='short_neutral',color='red')

plt.xlabel('new_polarity',fontsize =20)
plt.ylabel('Count',fontsize =20)
pyplot.legend(loc='upper right',fontsize=20)
plt.title('Neutral',fontsize =20)
```

```python
happy_short_df.to_csv("happy_short.csv")
neutral_short_df.to_csv("neutral_short.csv")
depression_short_df.to_csv("depression_short.csv")
depression_mid_df.to_csv("depression_mid.csv")
```
